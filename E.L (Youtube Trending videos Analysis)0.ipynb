```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel('/content/drive/MyDrive/DataSets/DSR Dataset (1).xlsx')
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
#Missing values

df['Location'].fillna('Unknown', inplace =True)
df['description'].fillna('',inplace =True)
df.isnull().sum()
```

```python
# Remove Duplicates
df.drop_duplicates(inplace =True)
df.duplicated().sum()
```

```python
# Convert trending date and publish time to datetime

df['trending_date'] = pd.to_datetime(df['trending_date'], format ='%y.%d.%m')
df['publish_time'] = pd.to_datetime(df['publish_time'])
```

```python
# Standardize Text Columns (lowercase)

df['title'] = df['title'].str.lower().str.strip()
df['tags'] = df['tags'].str.lower().str.strip()
df['description'] = df['description'].str.lower().str.strip()
```

```python
# Check for outliers in numerical columns

num_col = ['views','likes','dislikes','comment_count']
for col in num_col:
  Q1 = df[col].quantile(0.25)
  Q3 = df[col].quantile(0.75)
  IQR = Q3-Q1
  outliers = df[(df[col] <Q1 - 1.5 * IQR) | (df[col] > Q3 + 1.5* IQR)][col]
  print(f'Outliers in {col}: {len(outliers)}')
```

```python
# Assuming df is the cleaned DataFrame with sentiment and trending_duration columns
df.to_csv('cleaned_youtube_data_for_powerbi.csv', index=False)
print("Cleaned dataset exported as 'cleaned_youtube_data_for_powerbi.csv'")
```
